```python
import os
import json
import hashlib


CARPETA_DISCURSOS = "DiscursosOriginales"
ARCHIVO_SALIDA = "discursos-final-202373580-9-202473508-1.json"

model = SentenceTransformer("all-MiniLM-L6-v2")

def calcular_sha256(texto: str) -> str:
    h = hashlib.sha256()
    h.update(texto.encode("utf-8"))
    return h.hexdigest()

def procesar_discursos():
    documentos = []

    for nombre in sorted(os.listdir(CARPETA_DISCURSOS)):
        if not nombre.lower().endswith(".txt"):
            continue

        ruta_archivo = os.path.join(CARPETA_DISCURSOS, nombre)

        with open(ruta_archivo, "r", encoding="utf-8") as f:
            texto = f.read().strip()

        _id = calcular_sha256(texto)
        embedding = model.encode(texto).tolist()

        doc = {
            "_id": _id,
            "texto": texto,
            "embedding": embedding
        }

        documentos.append(doc)
        print(f"Procesado {nombre}")

    # Guardar como JSON en forma de ARRAY
    with open(ARCHIVO_SALIDA, "w", encoding="utf-8") as out:
        json.dump(documentos, out, ensure_ascii=False, indent=2)

    print(f"\nSe guardaron {len(documentos)} documentos en {ARCHIVO_SALIDA}")

if __name__ == "__main__":
    procesar_discursos()
```

```python
def verificar_consistencia_nodos():
    """Verifica que todos los nodos tengan los mismos datos"""
    print("\n=== VERIFICACIÓN DE CONSISTENCIA ENTRE NODOS ===")
    
    nodos = [
        "mongodb://mongo1:30001/?directConnection=true",
        "mongodb://mongo2:30002/?directConnection=true", 
        "mongodb://mongo3:30003/?directConnection=true"
    ]
    
    resultados = {}
    
    for i, url in enumerate(nodos, 1):
        try:
            client_nodo = MongoClient(url, serverSelectionTimeoutMS=5000)
            db_nodo = client_nodo["Politica"]
            coleccion_nodo = db_nodo["Discursos"]
            
            count = coleccion_nodo.count_documents({})
            
            sample_docs = list(coleccion_nodo.find().limit(3))
            sample_ids = [doc["_id"] for doc in sample_docs]
            
            resultados[f"nodo{i}"] = {
                "count": count,
                "sample_ids": sample_ids,
                "status": " CONECTADO"
            }
            
            client_nodo.close()
            
        except Exception as e:
            resultados[f"nodo{i}"] = {
                "count": 0,
                "sample_ids": [],
                "status": f" ERROR: {e}"
            }
    

    print("RESULTADOS POR NODO:")
    for nodo, info in resultados.items():
        print(f"  {nodo.upper()}: {info['status']}")
        print(f"    - Documentos: {info['count']}")
        if info['sample_ids']:
            print(f"    - Sample IDs: {info['sample_ids'][:2]}...")
    

    counts = [info['count'] for info in resultados.values() if info['count'] > 0]
    if len(set(counts)) == 1:
        print(f"\n CONSISTENCIA VERIFICADA: Todos los nodos tienen {counts[0]} documentos")
        return True
    else:
        print(f"\n INCONSISTENCIA DETECTADA: Conteos diferentes - {counts}")
        return False


print("Verificando consistencia después de insertar documentos...")
verificar_consistencia_nodos()
```

```python
def buscar_discursos_similares(consulta_texto, top_k=5):

    embedding_consulta = model.encode(consulta_texto).reshape(1, -1)
    
    documentos = list(coleccion.find({}))
    

    resultados = []
    for doc in documentos:
        embedding_doc = np.array(doc['embedding']).reshape(1, -1)
        similitud = cosine_similarity(embedding_consulta, embedding_doc)[0][0]
        
        resultados.append({
            '_id': doc['_id'],
            'texto': doc['texto'],
            'similitud': similitud
        })
    

    resultados_ordenados = sorted(resultados, key=lambda x: x['similitud'], reverse=True)[:top_k]
    
    return resultados_ordenados

consulta_ejemplo = "discurso sobre la democracia y la libertad"
resultados = buscar_discursos_similares(consulta_ejemplo)

print("Top 5 discursos más similares:")
for i, resultado in enumerate(resultados, 1):
    print(f"\n--- Resultado {i} ---")
    print(f"Similitud: {resultado['similitud']:.4f}")
    print(f"Texto: {resultado['texto']}")
    print(f"ID: {resultado['_id']}")
```

```python
print("=== PRUEBA DE BÚSQUEDA SEMÁNTICA ===")
consulta_ejemplo = "discurso sobre la democracia y la libertad"
resultados = buscar_discursos_similares(consulta_ejemplo)

print(f"Consulta: '{consulta_ejemplo}'")
print("\nTop 5 discursos más similares:")
for i, resultado in enumerate(resultados, 1):
    print(f"\n--- Resultado {i} (Similitud: {resultado['similitud']:.4f}) ---")
    print(f"Texto: {resultado['texto']}")
```

```python
def prueba_consistencia_operaciones():
    """Prueba que las operaciones se replican consistentemente"""
    print("\n=== PRUEBA DE CONSISTENCIA EN OPERACIONES CRUD ===\n")
    

    print("1. ESTADO INICIAL:")
    count_inicial = coleccion.count_documents({})
    print(f"   Documentos iniciales: {count_inicial}")

    print("\n2. INSERTANDO DOCUMENTO DE PRUEBA...")
    doc_prueba = {
        "_id": "test_consistencia_operaciones",
        "texto": "Documento para prueba de consistencia en operaciones CRUD",
        "embedding": model.encode("Documento prueba consistencia").tolist()
    }
    
    try:
        resultado_insercion = coleccion.insert_one(doc_prueba)
        print(f"   Documento insertado: {resultado_insercion.inserted_id}")
    except Exception as e:
        print(f"   Error en inserción: {e}")
        return
    
    print("\n3. VERIFICANDO REPLICACIÓN DE INSERCIÓN...")
    time.sleep(2)
    verificar_consistencia_nodos()
    
    print("\n4. ACTUALIZANDO DOCUMENTO...")
    try:
        resultado_actualizacion = coleccion.update_one(
            {"_id": "test_consistencia_operaciones"},
            {"$set": {"texto": "Documento ACTUALIZADO para prueba de consistencia"}}
        )
        print(f"   Documentos actualizados: {resultado_actualizacion.modified_count}")
    except Exception as e:
        print(f"   Error en actualización: {e}")
    
    print("\n5. VERIFICANDO REPLICACIÓN DE ACTUALIZACIÓN...")
    time.sleep(2)
    verificar_consistencia_nodos()
    
    print("\n7. ELIMINANDO DOCUMENTO...")
    try:
        resultado_eliminacion = coleccion.delete_one({"_id": "test_consistencia_operaciones"})
        print(f"   Documentos eliminados: {resultado_eliminacion.deleted_count}")
    except Exception as e:
        print(f"   Error en eliminación: {e}")
    
    print("\n8. VERIFICANDO REPLICACIÓN DE ELIMINACIÓN...")
    time.sleep(2)
    verificar_consistencia_nodos()
    
    count_final = coleccion.count_documents({})
    print(f"\n9. ESTADO FINAL: {count_final} documentos")
    
    if count_final == count_inicial:
        print("   CONSISTENCIA MANTENIDA: Conteo final igual al inicial")
    else:
        print("   INCONSISTENCIA: Conteo final diferente al inicial")

prueba_consistencia_operaciones()
```
